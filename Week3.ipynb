```python
# Recursive Function
def factorial(num):
    if num == 1:
        return 1
    else:
        result = num * factorial(num-1)
    return result
factorial(5)
```

```python
# Fibonaccis series:
def fibonacci(num):
  if num==0:
    return 0
  elif num ==1:
    return 1
  else:
    return fibonacci(num-1) + fibonacci(num-2)
print(fibonacci(4))
print(fibonacci(8))

def fib(num):
  a,b = 0,1
  for i in range(num):
    a,b = b,a+b
  return a
print(fib(4))
print(fib(8))



```

```python
# fact using for loop
num = 5
fact =1
for i in range(1,num+1):
  fact = fact * i
print(fact)

# Fact using while loop
n = 5
fact1 = 1
while n>=1:
  fact1 = fact1 * n
  n = n-1
print(fact1)
```

```python
# Filter Function
def isEven(x):
  if x%2 == 0:
    return True
  else:
    return False
l1 = [1,3,4,6,8,10]
l1 = list(filter(isEven,l1))
print(l1)

# With Lambda Function
l2 = [23,45,66,89,70,43,21,45,76]
l3 = list(filter(lambda x:x%2==0,l2))
print(l3)
```

```python
# Write a function to count vowels in string
def count_vowel(string):
  vowel='aieouAIEOU'
  count = 0
  for char in  string:
    if char in vowel:
      count = count + 1
  return count
count_vowel("Hello")

# To get vowels
s = "Hello Pramod"
vowels = [ch for ch in s if ch.lower() in "aieou"]
print(vowels)
```

```python
# To find largest number
l = [10,20,30,40,5]
def largest():
  largest = l[0]
  for num in l:
    if num > largest:
      largest = num
  return largest
largest()
```
