```python
# Aliasing
z = [1,2,3,4,5]
y = z
print("With Aliaing value changes reflect in other variable")
print(id(z))
print(id(y))
z[1] = 200

print(y)
print(z)

# Cloning
print("With Cloning value changes does not reflect in other variable")

x = [10,20,30,40,50]
y = x[:] # y = x.copy()
print(id(x))
print(id(y))
y[1] = 200
print(x)
print(y)
```

```python
# Copy and deepcopy
import copy
original = [1, 2, [3, 4]]
shallow = copy.copy(original)

shallow[2][0] = 999

print(original)
print(shallow)

original = [1, 2, [5, 6]]
deep = copy.deepcopy(original)

deep[2][0] = 888

print(original)
print(deep)
```
